```python
sigmavH11 = pd.read_csv('sigmavH11.csv',skiprows=4).astype(float)
sigmavW11 = pd.read_csv('sigmavW11.csv',skiprows=4).astype(float)
sigmavZ11 = pd.read_csv('sigmavZ11.csv',skiprows=4).astype(float)
sigmavt11 = pd.read_csv('sigmavt11.csv',skiprows=4).astype(float)
sigmavb11 = pd.read_csv('sigmavb11.csv',skiprows=4).astype(float)
```

```python
sigmaMO11 = pd.read_csv('/home/santiago/MicrOmegas/micromegas_6.2.3/micromegas_6.2.3/SingletDM/sigmav1.csv',skiprows=4)
weights = pd.read_csv('/home/santiago/MicrOmegas/micromegas_6.2.3/micromegas_6.2.3/SingletDM/weights.csv')
```

```python
weights
```

```python
all_parts = pd.DataFrame()
all_parts['T'] = sigmavH11['T']
all_parts['sigmav'] = sigmavH11['sigmav']*2.496430e-01+sigmavW11['sigmav']*4.992860e-01+sigmavZ11['sigmav']*2.496430e-01+sigmavb11['sigmav']*1.351331e-06+sigmavt11['sigmav']*3.380202e-07
all_parts['now'] = sigmavH11['sigmav']+sigmavW11['sigmav']+sigmavZ11['sigmav']+sigmavb11['sigmav']+sigmavt11['sigmav']
```

```python
plt.loglog(sigmavH11['T'],sigmavH11['sigmav'],label='h')
plt.loglog(sigmavW11['T'],sigmavW11['sigmav'],label='W')
plt.loglog(sigmavZ11['T'],sigmavZ11['sigmav'],label='Z')
plt.loglog(sigmavb11['T'],sigmavb11['sigmav'],label='b')
plt.loglog(sigmavt11['T'],sigmavt11['sigmav'],label='t')

plt.loglog(sigmaMO11['T'],sigmaMO11['sigmav'],label=r'$\mu\Omega$s')

plt.ylim(1e-42,1e-22)
plt.grid()
plt.legend()
```

```python
plt.loglog(sigmaMO11['T'],sigmaMO11['sigmav'],label=r'$\mu\Omega$s')
plt.loglog(all_parts['T'],all_parts['sigmav'],label='all')
plt.loglog(all_parts['T'],all_parts['now'],label='all (no weights)')
plt.grid()
plt.legend()
```
